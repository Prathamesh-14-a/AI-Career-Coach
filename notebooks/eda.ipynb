```python
# Standard job role distribution
role_count = df['Standardized_Job_Title'].value_counts()
```

```python
# bar plot 
plt.figure(figsize=(12,6))
role_count.plot(kind= 'bar' 
                , color= sns.color_palette('viridis', len(role_count)) ,
                edgecolor= 'black' , linewidth= 1.2)
plt.xlabel('Job Role' , fontsize= 12)
plt.ylabel('Count' , fontsize= 12)
plt.title('Top Job Role' , fontsize= 16)
plt.xticks(rotation=45)
plt.show()
```

Here we can see Data Analyst and Data Engineer job posting are highly dominating the market and Analytics Data Scientist job roles are moderately dominating the market

```python
# Location Distribution
location_count = df['Location'].value_counts()
```

```python
#bar plot for location distribution
location_count.head(10).plot(kind= 'bar'
                            , color= 'green',
                            alpha= 0.7
                            , edgecolor= 'black'
                            , linewidth= 1.2)
plt.xlabel('Location' , fontsize= 12)
plt.ylabel('Count' , fontsize= 12)
plt.title('Top Hiring Locations' , fontsize= 16)
plt.xticks(rotation=45)
plt.show()
```

Most of the jobs are posted for remote only and there are some locations like Benglore , Maharashtra with more than 10 job posting 

```python
# Experience Level Distribution
exp_counts = df['Experience_Years'].value_counts().sort_values()
```

```python
#bar plot for experience level distribution
exp_counts.plot(kind= 'barh'
                , color= 'orange',
                alpha= 0.7
                , edgecolor= 'black'
                , linewidth= 1.2)
plt.xlabel('Count' , fontsize= 12)
plt.ylabel('Experience Years' , fontsize= 12)
plt.title('Experience Required' , fontsize= 16)
plt.show()
```

Highest Number of Job Posting Require No Experince While Most Of the Jos Requires 2 or 1 year of Experince

```python
#Salary Analysis
df['salary_avg'].describe()
```

```python
sns.histplot(df['salary_avg'].dropna(), bins=30, kde=True, color='blue')
plt.xlabel('Average Salary', fontsize=12)
plt.title('Distribution of Average Salaries', fontsize=16)
plt.show()
```

The salaries of most of the jobs are lies between 2.2 LPA TO 5.5 LPA which might be for freshers and Highest Salaries like <10 LPA , <20 LPA , <30 LPA might be for senior level roles might require highes experience

```python
# Remote vs On-site Analysis
remote_count = df[df['Location'] == 'Remote'].shape[0]
onsite_count = df[df['Location'] != 'Remote'].shape[0]
```

```python
#pie 
labels = ['Remote', 'On-site']
sizes = [remote_count, onsite_count]
colors = ['lightblue', 'lightcoral']
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Remote vs On-site Job Distribution', fontsize=16)
plt.show()
```

There is minimal difference between number of job posting for remote and onsite but onsite dominated with 0.9 %
