```python

# INTERPRET MATCH SCORE
def interpret_match_score(
    match_score
):

    """
    Interpret candidate readiness
    based on match score.

    Parameters:
        match_score (float)

    Returns:
        str
    """
    # APPLICATION READY
   

    if match_score >= 80:

        return (

            "Your profile is strongly aligned "
            "with current market expectations "
            "for this role."
        )
    
    # MODERATELY ALIGNED
   
    elif match_score >= 60:

        return (

            "Your profile demonstrates "
            "moderate alignment with "
            "current market requirements."
        )

    # NEEDS UPSKILLING

    elif match_score >= 40:

        return (

            "Your profile shows partial "
            "alignment but requires "
            "additional upskilling."
        )

    # ------------------------------------------------
    # WEAK ALIGNMENT
    # ------------------------------------------------

    else:

        return (

            "Your current skill profile "
            "shows limited alignment "
            "with market expectations."
        )
```

```python
# ====================================================
# IDENTIFY STRENGTH AREAS
# ====================================================

def identify_strength_areas(
    matched_skills,
    recommendation_df
):

    """
    Generate strength insights
    from matched skills.

    Parameters:
        matched_skills (list)
        recommendation_df (pd.DataFrame)

    Returns:
        str
    """

    skill_insight_mapping = {
        "airflow": "Strong workflow orchestration skills for reliable data pipelines.",
        "artificial intelligence": "Experience with AI concepts and intelligent automation.",
        "aws": "Familiar with AWS cloud platforms for scalable analytics solutions.",
        "azure": "Familiar with Azure cloud tools for enterprise data workloads.",
        "big data": "Comfortable working with large-scale datasets and big data frameworks.",
        "business analysis": "Skilled at translating business needs into data-driven outcomes.",
        "dashboarding": "Strong dashboarding skills for effective data storytelling.",
        "data preprocessing": "Proficient at preparing and cleaning data for analysis.",
        "dbt": "Experience with dbt for transforming and documenting analytics data.",
        "etl": "Knowledgeable in ETL concepts for efficient data integration.",
        "excel": "Strong spreadsheet skills for analysis, reporting, and modeling.",
        "gcp": "Familiar with GCP tools for cloud-based data analytics.",
        "llm": "Understands large language models for advanced NLP workflows.",
        "market research": "Able to support market research with data insights and trends.",
        "nlp": "Skilled in natural language processing for text and language analysis.",
        "numpy": "Strong numerical computing skills using NumPy for data science workflows.",
        "pandas": "Strong data manipulation and analysis skills using pandas.",
        "power bi": "Experienced in Power BI for delivering interactive business dashboards.",
        "python": "Strong Python programming skills for analytics and automation.",
        "pytorch": "Knowledgeable in PyTorch for building machine learning models.",
        "r": "Strong R programming skills for statistical analysis and reporting.",
        "snowflake": "Familiar with Snowflake for scalable cloud data warehousing.",
        "spark": "Experience with Spark for processing large-scale distributed data.",
        "sql": "Strong SQL skills for querying and analyzing relational data.",
        "statistics": "Solid statistical foundations for data analysis and modeling.",
        "tableau": "Experienced in Tableau for creating compelling visualizations.",
        "tensorflow": "Knowledgeable in TensorFlow for building machine learning solutions.",
    }

    known_insights = []
    for skill in matched_skills:
        normalized_skill = skill.lower().strip()
        insight = skill_insight_mapping.get(normalized_skill)
        if insight:
            known_insights.append(insight)

    known_insights = sorted(dict.fromkeys(known_insights))
    if known_insights:
        if len(known_insights) == 1:
            return known_insights[0]
        return " ".join(known_insights)

    return (
        "Matched skills are recognized but do not match the core skill list "
        "used for insight generation."
    )
    
```
