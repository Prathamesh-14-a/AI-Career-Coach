```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r"d:\Startup\Project\ai-career-coach\data\processed\jobs_cleaned.csv")
df.columns
```

```python
df["combined_text"] = (
    df["Standardized_Job_Title"].fillna("") + " " +
    df["Job Description"].fillna("") + " " +
    df["Skills Required"].fillna("")
)
```

```python
df["Combined_text"][0]
```

```python
#removed punctuation and converted to lowercase
df["Combined_text"] = (df["Combined_text"].str.lower()
                       .apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', ' ', x)))
```

```python
#\xa0\xa0\r\n\xa0\r\n \r\n\r\n\r\n2\r\n\xa0\xa0\r\n\xa0\xa0\xa0\r\n\xa0\r\n\r\n\xa0\xa0\r\n\nIn removing these characters, we are essentially cleaning the text data by eliminating non-alphanumeric characters and extra whitespace. This can help in standardizing the text for further analysis, such as tokenization, keyword extraction, or machine learning tasks.
```

```python
df["Combined_text"] = df['Combined_text'].apply(lambda x : re.sub(r'[\r\n\xa0]+' , ' ' , x)).str.strip()
```

```python
df["Combined_text"][0]
```

```python
df["Combined_text"] = df['Combined_text'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
```

```python
df['Combined_text'] = df["Combined_text"].apply(lambda x: re.sub(r'\b\d+\b', '', x))
```

```python
df["Combined_text"][0]
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("punkt_tab")
```

```python
df['tokens'] = df["Combined_text"].apply(word_tokenize)
```

```python
df['tokens']
```

```python
import nltk
nltk.download("stopwords")
```

```python
def stopwords_removal(tokens):
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

df['filtered_tokens'] = df['tokens'].apply(stopwords_removal)
```

```python
df['filtered_tokens']
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
```

```python
df["lemmatized_tokens"] = df["filtered_tokens"].apply(lambda tokens: [lemmatizer.lemmatize(token , pos = "v") for token in tokens])
```

```python
df["lemmatized_tokens"][0]
```

```python
# skill matching 
df["processed_text"] = df["lemmatized_tokens"].apply(lambda x: " ".join(x))
```

```python
df["processed_text"]
```

```python
skills_list = ['python' ,'r','sql','java','scala','c++' ,
               'excel','power bi','tableau' ,'statistics',
               'data visualization','reporting','dashboarding' , 
               'mysql','postgresql','mongodb','snowflake','oracle' ,
               'etl','data warehousing','spark','hadoop','airflow','dbt',
               'machine learning','deep learning','nlp','tensorflow',
               'pytorch','scikit-learn','computer vision','llm' ,
               'aws','azure','gcp','databricks' ,'business analysis',
               'stakeholder management','kpi','forecasting','market research' ,
               'communication','problem solving','leadership','presentation',
               'pandas','numpy','matplotlib','seaborn','plotly','ggplot2','artificial intelligence','big data',
               'data mining','data cleaning', 'data preprocessing','data modeling','data analysis',
               'data storytelling', 'research','analysis','reporting','dashboard' ,'structured query language',
               'visualization','google sheets' , 'microsoft excel']

len(skills_list)
```

```python
def skill_extracter(text):
    extracted_skills = []

    for skill in skills_list:
        
        if re.search(r'\b' + re.escape(skill) + r'\b', text):
            extracted_skills.append(skill)

    return list(set(extracted_skills))


df["extracted_skills"] = df["processed_text"].apply(skill_extracter)
```

```python
df["skill_count"] = df["extracted_skills"].apply(len)
```

```python
all_skills = [
    skill 
    for skill_list in df['extracted_skills'] 
    for skill in skill_list
]
len(all_skills)
```

```python
# count skill frequency
from collections import Counter
skill_counts = Counter(all_skills)
```

```python
normalization_dict = {
    "microsoft excel": "excel",
    "structured query language": "sql",
    "visualization": "data visualization",
    "dashboard": "dashboarding"
}
df['extracted_skills'] = df['extracted_skills'].apply(lambda skills: [normalization_dict.get(skill, skill) for skill in skills])
```

```python
role_skill_map = df.groupby("Standardized_Job_Title")["extracted_skills"].sum()
```

```python
role_skill_map
```

```python
role_skill_data = []

for role in df["Standardized_Job_Title"].unique():
    
    role_df = df[df["Standardized_Job_Title"] == role]
    
    role_skill = [
        skill
        for skills_list in role_df["extracted_skills"]
        for skill in skills_list
    ]

    role_skill_count = Counter(role_skill)

    for skill , count in role_skill_count.items():
        role_skill_data.append({
            "Role": role,
            "Skill": skill,
            "Count": count
        })
```

```python
role_skill_df = pd.DataFrame(role_skill_data)
role_skill_df.reset_index(inplace=True, drop=True)
```

```python
role_skill_df = role_skill_df.sort_values(by=["Role" , "Count"] , ascending=[True, False])
```

```python
role_skill_df.head()
```

```python
role_skill_matrix = role_skill_df.pivot_table(
    index = 'Role' , 
    columns = 'Skill' ,
    values = 'Count' , 
    fill_value = 0
)
```
