```python
import pdfplumber
import re
import pandas as pd
```

```python
pdf_path = r"d:\Startup\Project\ai-career-coach\data\resume\Pratham_resume.pdf"
pages_text = []
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        page_text = page.extract_text()

        if page_text :
            pages_text.append(page_text)
        
full_text = " ".join(pages_text)
full_text
```

```python
full_text = full_text.lower()
full_text = re.sub(r'[^a-zA-Z0-9\s]', ' ' , full_text)
full_text = re.sub(r'[\r\n\xa0]+' , ' ' , full_text)
full_text = re.sub(r'\s+', ' ', full_text)
full_text = re.sub(r'\b\d+\b', '', full_text)
full_text = re.sub(r"[•★►▪]", " ", full_text)
full_text
```

```python
technical_skills = [
    
    'python', 'r', 'sql', 
    'excel' 'microsoft excel',
    'power bi', 'tableau',

    'etl', 'spark', 'hadoop',
    'airflow', 'dbt',

    'machine learning',
    'tensorflow', 'pytorch', 'scikit-learn',

    'aws', 'azure', 'gcp', 

    'pandas', 'numpy', 'matplotlib',
    'seaborn', 'plotly', 'ggplot2',

    ]
```

```python
def extract_skills(resume_text: str, skills_list: list) -> list:
    skill_set = set()
    for skill in skills_list:
        escape_skill = re.escape(skill)

        pattern = rf'\b{escape_skill}\b'

        if re.search(pattern , resume_text):
            skill_set.add(skill)

    return sorted(skill_set)

extract_skills(full_text , technical_skills)
```
