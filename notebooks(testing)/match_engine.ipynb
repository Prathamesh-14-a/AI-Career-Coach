```python
import pandas as pd
```

```python
resume_skills = [
    "sql",
    "excel",
    "power bi",
    "tableau",
    "python"
]
resume_skills = [
    skill.lower().strip()
    for skill in resume_skills
]
resume_skills = list(set(resume_skills))
```

```python
print(resume_skills)
```

```python
df = pd.read_csv(r'd:\startup\Project\ai-career-coach\data\processed\top_skills_by_role.csv')
```

```python
print(
    df[
        "Standardized_Job_Title"
    ]
    .unique()
)
```

```python
# ---------------------------------------------------
# TARGET ROLE
# ---------------------------------------------------

target_role = "Data Analyst"

# ---------------------------------------------------
# NORMALIZE TARGET ROLE
# ---------------------------------------------------

target_role = (
    target_role
    .strip()
)

# ---------------------------------------------------
# FILTER ROLE BENCHMARK
# ---------------------------------------------------

role_df = (
    df[
        df[
            "Standardized_Job_Title"
        ]
        .str.strip()
        == target_role
    ]
)

# ---------------------------------------------------
# PREVIEW
# ---------------------------------------------------

print("\nTarget Role:")
print(target_role)

print("\nRole Benchmark Preview:")

print(
    role_df.head(10)
)

# ---------------------------------------------------
# VALIDATION
# ---------------------------------------------------

print("\nShape:")
print(role_df.shape)

print("\nUnique Roles:")

print(
    role_df[
        "Standardized_Job_Title"
    ].unique()
)
```

```python
role_skills = role_df['Skill'].tolist()
```

```python
resume_set = set(resume_skills)

role_set = set(role_skills)
```

```python
print(type(resume_set))
print(type(role_set))
```

```python
matched_skills = sorted(
    resume_set & role_set
)
```

```python
print("\nMatched Skills:")

print(matched_skills)
```

```python

missing_skills = sorted(
    role_set - resume_set
)
```

```python
missing_skills
```

```python
role_skill_count = len(role_set)
matched_skills_count = len(matched_skills)
```

```python
if len(role_set) > 0:

    match_score = (
        len(matched_skills)
        / len(role_set)
    ) * 100

else:

    match_score = 0
```

```python
match_score = round(match_score , 2)
print(f"\nMatch Score: {match_score}%")
```

```python
match_result = {

    "Role": target_role,

    "Match Score": match_score,

    "Matched Skills": matched_skills,

    "Missing Skills": missing_skills,

    "Matched Skill Count": len(
        matched_skills
    ),

    "Missing Skill Count": len(
        missing_skills
    )
}
```
