```python
import re


# ============================================
# SKILL STANDARDIZATION MAPPING
# ============================================

SKILL_MAPPING = {

    # Programming Languages
    "py": "python",
    "python3": "python",
    "r": "r programming",
    "r language": "r programming",
    "c++": "cpp",
    "c plus plus": "cpp",
    "js": "javascript",

    # Machine Learning / AI
    "ml": "machine learning",
    "dl": "deep learning",
    "nlp": "natural language processing",
    "ai": "artificial intelligence",

    # Databases
    "postgres": "postgresql",
    "postgre": "postgresql",
    "mongo": "mongodb",

    # Cloud
    "aws cloud": "aws",
    "amazon web services": "aws",
    "gcp": "google cloud platform",

    # Visualization
    "powerbi": "power bi",
    "ms excel": "excel",
    "microsoft excel": "excel",

    # Big Data
    "hdfs": "hadoop",
}


# ============================================
# CLEAN SINGLE SKILL
# ============================================

def clean_single_skill(skill):
    """
    Clean and standardize one skill token.

    Parameters:
        skill (str)

    Returns:
        str
    """

    if pd.isna(skill):
        return ""

    # lowercase
    skill = str(skill).lower().strip()

    # remove quotes
    skill = skill.replace("'", "")
    skill = skill.replace('"', "")

    # remove extra spaces
    skill = re.sub(r"\s+", " ", skill)

    # remove special characters except +
    skill = re.sub(r"[^a-zA-Z0-9+# ]", "", skill)

    # apply mapping
    skill = SKILL_MAPPING.get(skill, skill)

    return skill


# ============================================
# CLEAN COMPLETE SKILL STRING
# ============================================

def clean_skills(skills_text):
    """
    Clean complete skills column.

    Parameters:
        skills_text (str)

    Returns:
        str
    """

    if pd.isna(skills_text):
        return ""

    # lowercase
    skills_text = str(skills_text).lower()

    # split skills
    raw_skills = re.split(r",|\||/|;", skills_text)

    cleaned_skills = []

    for skill in raw_skills:

        cleaned_skill = clean_single_skill(skill)

        # skip empty tokens
        if cleaned_skill == "":
            continue

        # skip very short noisy tokens
        if len(cleaned_skill) <= 1:
            continue

        cleaned_skills.append(cleaned_skill)

    # remove duplicates
    cleaned_skills = list(set(cleaned_skills))

    # sort for consistency
    cleaned_skills.sort()

    # convert back to string
    return ", ".join(cleaned_skills)


# ============================================
# APPLY TO DATAFRAME
# ============================================

df["Skills Required"] = df["Skills Required"].apply(clean_skills)
```
