```python
def clean_location(location):

    if pd.isna(location):
        return "Unknown"

    location = str(location)

    # remove +more
    location = re.sub(r"\+.*more", "", location)

    # remove tabs/newlines
    location = re.sub(r"\s+", " ", location).strip()

    # split multi-city
    location = location.split(",")[0].strip()

    # mapping
    location_map = {

        "Bangalore / Bengaluru": "Bengaluru",
        "Bengaluru/Bangalore": "Bengaluru",
        "Bangalore Rural": "Bengaluru",
        "Bangalore/Bengaluru": "Bengaluru",

        "Mohali/SAS Nagar" : "Mohali" ,

        "Trivandrum/Thiruvananthapuram" : "Thiruvananthapuram" ,

        "Delhi/Ncr": "Delhi NCR",
        "New Delhi": "Delhi NCR",
        "Delhi / NCR": "Delhi NCR",

        "Gurgaon / Gurugram": "Gurugram",
        "Gurgaon/Gurugram": "Gurugram",

        "Mumbai suburbs": "Mumbai",
        "Mumbai Suburbs": "Mumbai",
        "Navi Mumbai": "Mumbai",

        "Hyderabad / Secunderabad": "Hyderabad",
        "Hyderabad/Secunderabad": "Hyderabad",

        "Mangaluru/Mangalore" : "Mangalore" , 

        "Panaji/Panjim" : "Panaji" ,

        "Calicut/Kozhikode" : "Calicut" , 

        "Kochi/Cochin" : "Kochi" ,
        "Cochin/Ernakulam/Kochi" : "Kochi",

        "Remote": "Remote",

        "India": "Remote"
    }

    return location_map.get(location, location)
```

```python
def scrape_data(role, num_pages=1, base_url="https://www.ambitionbox.com/jobs"):
    """Scrape job listings for a given role and return a dataframe.

    role: string like 'data-analyst', 'data scientist', or 'machine-learning-engineer'
    num_pages: number of pagination pages to scrape
    """
    def normalize_role(role_text):
        text = str(role_text).strip().lower()
        return text.replace(" ", "-").replace("/", "-").replace("_", "-")

    role_slug = normalize_role(role)
    headers = {
        "User-Agent": "Mozilla/5.0",
        "Accept-Language": "en-US,en;q=0.9"
    }

    rows = []

    for page in range(1, num_pages + 1):
        url = f"{base_url}/{role_slug}-jobs-prf?page={page}"
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()
        except requests.exceptions.Timeout:
            print(f"Page {page}: request timed out for role {role}")
            continue
        except requests.exceptions.RequestException as e:
            print(f"Page {page}: request failed for role {role}: {e}")
            continue

        soup = BeautifulSoup(response.text, "html.parser")
        cards = soup.find_all("div", class_=lambda x: x and "jobInfoCard" in x)

        if not cards:
            print(f"Page {page}: no job cards found for role {role} using URL {url}")

        for card in cards:
            title_tag = card.find("a", class_=lambda x: x and "title" in x)
            title = title_tag.text.strip() if title_tag else None

            exp_div = card.find("div", title=lambda x: x and "year" in x.lower())
            experience = exp_div.find("p").text.strip() if exp_div and exp_div.find("p") else None

            loc_div = card.find("div", class_=lambda x: x and "loc" in x)
            location = loc_div.find("p").text.strip() if loc_div and loc_div.find("p") else None

            salary_div = card.find("div", title=lambda x: x and "L/yr" in x)
            salary = salary_div.find_next("p").text.strip() if salary_div else None

            skill_divs = soup.find_all(
            "div",
            title=True
            )
            for div in skill_divs:
                title = div.get("title")
            # skills usually contain commas
                if title and "," in title:
                    skills = list(title.split(","))

        rows.append({
                "Role": role,
                "Job Title": title,
                "Experience": experience,
                "Location": location,
                "Salary": salary,
                "Skills": skills
            })

            # Extract skills from page-level divs with title attribute containing commas
        



    return pd.DataFrame(rows)


def scrape_roles(roles, num_pages=1):
    """Scrape multiple roles and return a dict of dataframes."""
    return {role: scrape_data(role, num_pages=num_pages) for role in roles}
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9"
}


def scrape_jobs(role, pages):

    all_jobs = []

    # convert role to url format
    role_url = role.lower().replace(" ", "-")

    for page in range(1, pages + 1):

        url = f"https://www.ambitionbox.com/jobs/{role_url}-jobs-prf?page={page}"

        print(f"Scraping Page {page}")

        try:
            response = requests.get(
                url,
                headers=headers,
                timeout=10
            )

            response.raise_for_status()

        except requests.exceptions.Timeout:
            print("Request timed out")
            continue

        except requests.exceptions.RequestException as e:
            print("Request failed:", e)
            continue

        soup = BeautifulSoup(response.text, "html.parser")

        cards = soup.find_all("div", class_="jobInfoCard")

        for card in cards:

            # TITLE
            title_tag = card.find("a", class_="title noclick")

            job_title = (
                title_tag.text.strip()
                if title_tag else None
            )

            # EXPERIENCE
            exp_div = card.find(
                "div",
                title=lambda x: x and "year" in x.lower()
            )

            experience = (
                exp_div.get("title")
                if exp_div else None
            )

            # LOCATION
            loc_div = card.find(
                "div",
                class_=lambda x: x and "loc" in x
            )

            location = (
                loc_div.get("title")
                if loc_div else None
            )

            # SALARY
            salary_div = card.find(
                "div",
                title=lambda x: x and "L/yr" in x
            )

            salary = (
                salary_div.get("title")
                if salary_div else None
            )

            # SKILLS
            skills = None

            skill_divs = card.find_all("div", title=True)

            for div in skill_divs:

                skill_title = div.get("title")

                # skills usually contain commas
                if skill_title and "," in skill_title:

                    skills = [
                        skill.strip()
                        for skill in skill_title.split(",")
                    ]

                    break

            # STORE DATA
            all_jobs.append({
                "role": role,
                "job_title": job_title,
                "experience": experience,
                "location": location,
                "salary": salary,
                "skills": skills
            })

    # dataframe
    df = pd.DataFrame(all_jobs)

    return df
```
