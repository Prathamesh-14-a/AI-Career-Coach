```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://internshala.com/jobs/analytics,data-science,machine-learning-jobs/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive"
}
response = requests.get(url , headers=headers)
soup = BeautifulSoup(response.text , "html.parser")
print(response.status_code)
```

```python
soup.find_all("a" ,class_ = "job-title-href")[0].text.strip()
```

jobs extraction

```python
job_titles = []

for titile in soup.find_all("a" , class_ = "job-title-href"):
    job_titles.append(titile.text.strip())
```

```python
print(job_titles )
print(len(job_titles))
```

```python
relevant_keywords = [
    "data",
    "analyst",
    "scientist",
    "machine learning",
    "ai",
    "analytics",
    "business analyst",
    "etl"
]
```

```python
filtered_jobs = []

for titles in job_titles:
    lower_titles = titles.lower()
    if any(keywords in lower_titles for keywords in relevant_keywords):
        filtered_jobs.append(titles)

print(filtered_jobs)
print("relevant jobs : ",
      len(filtered_jobs))
```

location extractions

```python
locations = []
for job in soup.find_all("div", class_="individual_internship"):
    location_tag = job.select_one("p.row-1-item.locations span a")
    
    if location_tag:
        locations.append(location_tag.get_text(strip=True))

print(locations[:5])
print(len(locations))
```

salary extraction

```python
salary = []
for sal in soup.find_all("div" , class_= "row-1-item"):
    sal_tag = sal.select_one("span.desktop")

    if sal_tag:
        salary.append(sal_tag.get_text(strip=True))    
print(salary[:5])
print(len(salary))
```

Experience Extraction

```python
experience_list = []

for job in soup.find_all("div", class_="individual_internship_details"):
    exp_tag = job.select_one("i.ic-16-briefcase + span")
    
    experience = exp_tag.get_text(strip=True)
    
    experience_list.append(experience)

print(experience_list)
```

```python
print(len(experience_list))
```

skills extraction

```python


skills_required= []

# Loop through each job card
for job in soup.find_all("div", class_="individual_internship"):
    
    # Find all skill tags inside that specific job
    skill_tags = job.select("div.job_skills div.job_skill")
    
    # Extract skill names into list
    skills = [skill.get_text(strip=True) for skill in skill_tags]
    
    # Store one job's skills as sublist
    skills_required.append(skills)

print(skills_required)
print(len(skills_required))
```

job description extractor 

```python
job_description = []

for des in soup.find_all("div", class_="individual_internship_details"):
    des_tag = des.select_one("div.about_job div.text")
    
    description = des_tag.get_text(strip=True)
    
    job_description.append(description)

print(job_description)
print(len(job_description))
```

posted Dates Extraction


```python
posted_dates = []

for job in soup.find_all("div", class_="internship_meta"):
    date = job.select_one("div.detail-row-2 div.color-labels span")
    posted_dates.append(date.text.strip() if date else "No date found")
```

job link

```python
job_links = []
for link in soup.find_all("div", class_="internship_meta"):
    link_tag = link.select_one('a.job-title-href')
    if link_tag:
        job_links.append(link_tag['href'])
print(job_links[:5])
```
