```python
def clean_location(location):

    if pd.isna(location):
        return "Unknown"

    location = str(location)

    # remove +more
    location = re.sub(r"\+.*more", "", location)

    # remove tabs/newlines
    location = re.sub(r"\s+", " ", location).strip()

    # split multi-city
    location = location.split(",")[0].strip()

    # mapping
    location_map = {

        "Bangalore / Bengaluru": "Bengaluru",
        "Bengaluru/Bangalore": "Bengaluru",
        "Bangalore Rural": "Bengaluru",
        "Bangalore/Bengaluru": "Bengaluru",

        "Mohali/SAS Nagar" : "Mohali" ,

        "Trivandrum/Thiruvananthapuram" : "Thiruvananthapuram" ,

        "Delhi/Ncr": "Delhi NCR",
        "New Delhi": "Delhi NCR",
        "Delhi / NCR": "Delhi NCR",

        "Gurgaon / Gurugram": "Gurugram",
        "Gurgaon/Gurugram": "Gurugram",

        "Mumbai suburbs": "Mumbai",
        "Mumbai Suburbs": "Mumbai",
        "Navi Mumbai": "Mumbai",

        "Hyderabad / Secunderabad": "Hyderabad",
        "Hyderabad/Secunderabad": "Hyderabad",

        "Mangaluru/Mangalore" : "Mangalore" , 

        "Panaji/Panjim" : "Panaji" ,

        "Calicut/Kozhikode" : "Calicut" , 

        "Kochi/Cochin" : "Kochi" ,
        "Cochin/Ernakulam/Kochi" : "Kochi",

        "Remote": "Remote",

        "India": "Remote"
    }

    return location_map.get(location, location)
```

```python


def extract_avg_exp(exp):

    if pd.isna(exp):
        return None

    exp = str(exp).lower().strip()

    # extract all numbers
    nums = re.findall(r'\d+', exp)

    if len(nums) >= 2:
        min_exp = int(nums[0])
        max_exp = int(nums[1])

        return (min_exp + max_exp) / 2

    elif len(nums) == 1:
        return int(nums[0])

    return None


data2['avg_experience'] = data2['experience'].apply(extract_avg_exp)
```

```python
def flatten_nested_list_v2(item):
    """Recursively flatten nested lists"""
    if not isinstance(item, list):
        return item
    
    result = []
    for element in item:
        if isinstance(element, list):
            # Recursively flatten nested lists
            flattened = flatten_nested_list_v2(element)
            if isinstance(flattened, list):
                result.extend(flattened)
            else:
                result.append(flattened)
        else:
            result.append(element)
    return result

# Apply the improved flattening function
salary_final_data['Skills Required'] = salary_final_data['Skills Required'].apply(flatten_nested_list_v2)

# Verify the result
print("Sample after flattening:")
print(salary_final_data[['Standardized_Job_Title', 'Skills Required']].head(10))
```

```python
import ast

def clean_skills_column(item):
    """Clean and flatten the Skills Required column"""
    if not isinstance(item, list):
        return item
    
    result = []
    for element in item:
        if isinstance(element, str):
            # Remove leading/trailing brackets and quotes
            cleaned = element.strip()
            if cleaned.startswith("["):
                try:
                    # Try to parse as a list
                    parsed = ast.literal_eval(cleaned)
                    if isinstance(parsed, list):
                        result.extend(parsed)
                    else:
                        result.append(parsed)
                except:
                    # If parsing fails, clean up quotes and add
                    cleaned = cleaned.strip("'\"[]")
                    if cleaned:
                        result.append(cleaned)
            else:
                # Remove quotes and add
                cleaned = cleaned.strip("'\"")
                if cleaned:
                    result.append(cleaned)
        elif isinstance(element, list):
            result.extend(element)
        else:
            result.append(element)
    
    return result

# Apply the cleaning function
salary_final_data['Skills Required'] = salary_final_data['Skills Required'].apply(clean_skills_column)

# Verify the result
print("Sample after cleaning:")
print(salary_final_data[['Standardized_Job_Title', 'Skills Required']].head(10))
```

```python
import re

def clean_skills_final(item):
    """Final cleaning of Skills Required column - handles all edge cases"""
    if not isinstance(item, list):
        return item
    
    result = []
    for element in item:
        if isinstance(element, str):
            # Remove all leading/trailing quotes, brackets, and spaces
            cleaned = element.strip().strip("'\"[]()").strip()
            if cleaned:
                result.append(cleaned)
        elif isinstance(element, list):
            # Recursively handle nested lists
            nested_result = clean_skills_final(element)
            if isinstance(nested_result, list):
                result.extend(nested_result)
            else:
                result.append(nested_result)
        else:
            result.append(element)
    
    return result

# Apply the final cleaning function
salary_final_data['Skills Required'] = salary_final_data['Skills Required'].apply(clean_skills_final)

# Verify the result
print("Sample after final cleaning:")
for i in range(min(5, len(salary_final_data))):
    print(f"Row {i}: {salary_final_data['Skills Required'].iloc[i][:3]}...")  # Show first 3 skills
    
print("\nFull samples:")
print(salary_final_data[['Standardized_Job_Title', 'Skills Required']].head(10))
```
