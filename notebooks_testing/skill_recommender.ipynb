```python
import pandas as pd
```

```python
df = pd.read_csv(r"d:/Startup/Project/ai-career-coach/data/processed/data_analyst_batch_results.csv")
```

```python
df1 = pd.read_csv(r"d:/Startup/Project/ai-career-coach/data/processed/top_skills_by_role.csv")
```

```python
df1
```

```python
missing_skill_str = df.loc[0, "Missing Skills"]
```

```python
missing_skills = missing_skill_str.split(',')
```

```python
missing_skills = [
    skill.lower().strip()
    for skill in missing_skills
]
```

```python
missing_skills
```

```python
df1["Standardized_Job_Title"] = (

        df1[
            "Standardized_Job_Title"
        ]

        .str.lower()
        .str.strip()
    )
```

```python

df1["Skill"] = (

      df1["Skill"]

        .str.lower()
        .str.strip()
    )
```

```python
target_role = (
        "Data Analyst"
        .lower()
        .strip()
    )
```

```python
filtered_df =df1[

    (df1['Standardized_Job_Title'] == 'data analyst')
    &
        (df1["Skill"]
        .isin(missing_skills))
    ]
```

```python
filtered_df
```

```python
prioritized_df = (

        filtered_df

        .sort_values(
            by="Count",
            ascending=False
        )

        .reset_index(drop=True)
    )
```

```python
prioritized_df
```

```python
high_priority = prioritized_df['Count'].max()
low_priority = prioritized_df['Count'].min()
```

```python
def get_priority(count):

    if count >= 15:

        return "High"

    elif count >= 7:

        return "Medium"

    else:

        return "Low"
```

```python
prioritized_df['priority'] = prioritized_df['Count'].apply(get_priority)
```

```python
prioritized_df
```

```python
roadmap = (

        prioritized_df["Skill"]

        .tolist()
    )
```

```python
roadmap
```

```python
print("\nRecommended Learning Path:\n")

for index, skill in enumerate(
    roadmap,
    start=1
):

    print(f"{index}. {skill.title()}")
```
